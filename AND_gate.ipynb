```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import optimizers
```

```python
# Input data: x1 and x2
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Output data
Y_and = np.array([[0], [0], [0], [1]])  # AND operation
Y_or = np.array([[0], [1], [1], [1]])   # OR operation
Y_xor = np.array([[0], [1], [1], [0]])  # XOR operation
```

```python
# Create the model
model_and = Sequential()
model_and.add(Dense(4, input_dim=2, activation='relu'))
model_and.add(Dense(1, activation='sigmoid'))

# Compile the model
model_and.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model_and.fit(X, Y_and, epochs=100, verbose=1)
```

```python
# Evaluate the model
loss_and, accuracy_and = model_and.evaluate(X, Y_and)
print(f'AND Operation - Loss: {loss_and}, Accuracy: {accuracy_and}')
```

```python
# Create the model
model_or = Sequential()
model_or.add(Dense(4, input_dim=2, activation='relu'))
model_or.add(Dense(1, activation='sigmoid'))

# Compile the model
model_or.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model_or.fit(X, Y_or, epochs=100, verbose=1)

# Evaluate the model
loss_or, accuracy_or = model_or.evaluate(X, Y_or)
print(f'OR Operation - Loss: {loss_or}, Accuracy: {accuracy_or}')
```

```python
# Create the model
model_xor = Sequential()
model_xor.add(Dense(4, input_dim=2, activation='relu'))
model_xor.add(Dense(1, activation='sigmoid'))

# Compile the model
model_xor.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model_xor.fit(X, Y_xor, epochs=100, verbose=1)

# Evaluate the model
loss_xor, accuracy_xor = model_xor.evaluate(X, Y_xor)
print(f'XOR Operation - Loss: {loss_xor}, Accuracy: {accuracy_xor}')
```

```python
# Save the model for AND operation
model_and.save("model_and.keras")

# Save the model for OR operation
model_or.save("model_or.keras")

# Save the model for XOR operation
model_xor.save("model_xor.keras")
```
